```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
tamanho = np.array([6,8,10,14,18,20,25,30])
preco = np.array([7,9.2,15,18.8,23.58,25.7,30,35.9])
```

```python
plt.scatter(tamanho,preco)
plt.xlabel('Tamanho da pizza')
plt.ylabel('Preço da pizza em relação ao diâmetro')
plt.title('Preço da pizza')
plt.show()
```

```python
model = LinearRegression()
tamanho = tamanho.reshape(-1, 1)
model.fit(tamanho, preco)
```

```python
preco_pred = model.predict(tamanho)
mse= mean_squared_error(preco, preco_pred)
r2 = r2_score(preco, preco_pred)
print(f'Erro médio quadrático: {mse}')
print(f'Coeficiente de determinação (R²): {r2*100:.2f}%')
```

```python
tamanho_novo = np.array([24])
tamanho_novo = tamanho_novo.reshape(-1, 1)
preco_novo = model.predict(tamanho_novo)
print(tamanho_novo, preco_novo)
```

```python
plt.scatter(tamanho, preco, color='blue', label='Dados reais')
plt.plot(tamanho, preco_pred, color='red', label='Linha de regressão')

plt.scatter(tamanho_novo, preco_novo, color='green', label='Novo ponto', s=100)

plt.xlabel('Tamanho da pizza')
plt.ylabel('Preço da pizza')
plt.title('Preço da pizza em relação ao diâmetro')
plt.legend()
plt.show()
```
